```python
# Install required packages
# !pip install tiatoolbox matplotlib numpy

from tiatoolbox.wsicore.wsireader import WSIReader
from tiatoolbox.tools import stainnorm
import matplotlib.pyplot as plt
import numpy as np

# Paths to WSIs
target_slide_path = "/tank/WSI_data/Ovarian_WSIs/MSK-IMPACT/hne/mskcc/394140.svs"
test_slide_path = "/tank/WSI_data/Ovarian_WSIs/BRCA_blind_test/slides/MP1467_23-1-1-1.mrxs"

# Load WSIs
target_wsi = WSIReader.open(target_slide_path)
test_wsi = WSIReader.open(test_slide_path)

# Read the entire slides at level 0 (highest resolution)
target_wsi_full = np.array(target_wsi.read_region(location=[0, 0], level=2, size=target_wsi.info.slide_dimensions))
test_wsi_full = np.array(test_wsi.read_region(location=[0, 0], level=2, size=test_wsi.info.slide_dimensions))

# Ensure the arrays are writable
target_wsi_full = target_wsi_full.copy()
test_wsi_full = test_wsi_full.copy()

# Initialize the stain normalizer
normalizer = stainnorm.VahadaneNormalizer()
normalizer.fit(target_wsi_full)  # Fit the normalizer using the entire target WSI

# Normalize the entire test WSI
normalized_wsi = normalizer.transform(test_wsi_full)

# Plotting
plt.figure(figsize=(15, 10))

plt.subplot(1, 3, 1)
plt.imshow(target_wsi_full)
plt.title("Target WSI")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(test_wsi_full)
plt.title("Test WSI (Pre-Normalization)")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(normalized_wsi)
plt.title("Test WSI (Post-Normalization)")
plt.axis("off")

plt.tight_layout()
plt.show()
```
